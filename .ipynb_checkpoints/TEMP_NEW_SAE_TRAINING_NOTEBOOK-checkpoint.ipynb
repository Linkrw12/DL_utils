```python
class EncodedSeqDataset(Dataset):
    
    def __init__(self, encoded_seqs):
        self.seqs = encoded_seqs
        
    def __len__(self):
        return len(self.seqs)
    
    def __getitem__(self, idx):
        return self.seqs[idx], self.seqs[idx]
```

```python
import matplotlib.pyplot as plt
import seaborn as sbn

def plot_train_val_loss(train_loss, val_loss,
                        figname = None):
    """
    As the name implies...Takes the train_loss and val_loss
    outputs from the training functions and plots the change
    in loss over each epoch
    """
    
    fig, ax = plt.subplots(1,1)
    
    ax.plot(train_loss)
    ax.plot(val_loss)
    ax.legend(['train', 'val'], loc='upper right')
    
    ax.set_xlabel("Epoch")
    ax.set_ylabel("Loss")
    
    sbn.despine()
    
    if figname:
        
        # Creates title based on the path name
        file_split = figname.split('/')
        
        sae   = file_split[2]
        fold  = file_split[1].split('_')[-1]
        cycle = file_split[3].split('_')[-1]
        
        formater = {'rna_sae': "RNA SAE",
                    'prot_sae': "Protein SAE",
                    'prot_struct_sae': "Protein Structure SAE",
                    'rna_struct_sae': "RNA Structure SAE"}
        
        ax.set_title(f"Fold {fold} {formater[sae]} Cycle {cycle} Train-Val Loss")
        plt.savefig(figname, dpi = 300)
```

```python
import os

def find_best_weights(weight_dir):
    """
    In a weight directory, returns the file with the largest
    epoch. This returns the best performing model for that 
    training session
    
    This looks for the last instance of an underscore and the
    last instance of a period and collects their index. Then use
    indexing to isolate the epoch number. We then search for the
    largest number in the list and return its index. This index is
    then used to find the file with the best performing model weights
    """
    
    weights = os.listdir(weight_dir)
    weights = [file for file in weights if file[-4:] == ".hdf"]
        
    epochs  = [int(file[file.rfind("_") + 1 : file.rfind(".")]) for file in weights]
    
    max_idx = epochs.index( max(epochs) )
        
    return weights[max_idx]
```
